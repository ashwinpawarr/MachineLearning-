Working with dates

```python
date["date"] = pd.to_datetime(date["date"])
```

```python
date.info()
```

1.Extract year

```python
date["date_year"] = date['date'].dt.year

date.sample(5)
```

2.Extract month

```python
date["month"] = date['date'].dt.month
date.sample(5)
```

```python
date['month_name'] = date['date'].dt.month_name()
date.sample(5)
```

```python
date["date_day"] = date['date'].dt.day
date.sample(5)
```

```python
date["day_of_week"] = date["date"].dt.day_of_week
date.sample(5)
```

```python
date["date_day_name"] = date['date'].dt.day_name()

date.drop(columns=['product_id','city_id','orders','month']).head()
```

```python
#is this weekend?
date['date_is_weekend'] = np.where(date['date_day_name'].isin(["Saturday",'Sunday']), 1,0)

date.drop(columns=['product_id','city_id','orders','month']).head()
```

 Extract week of the year

```python
date['date_week'] = date["date"].dt.weekday

date.drop(columns=['product_id','city_id','orders','month']).head()
```

```python
date["Quarter"] = date["date"].dt.quarter

date.drop(columns=['product_id','city_id','orders','month']).head()
```

```python
date['semester'] = np.where(date['Quarter'].isin([1,2]), 1,2)

date.drop(columns=['product_id','city_id','orders','month']).head()
```

```python
import datetime

today = datetime.datetime.today()

today
```

```python
today - date['date']
```

```python
(today - date['date']).dt.days
```

```python
# Months passed

np.round((today -date['date']) / np.timedelta64(1, "m"),0)
```
